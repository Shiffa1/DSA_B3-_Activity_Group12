### 4.Carry out correlation analysis on the dataset and find the correlation factor between .What do we infer from this correlation value?
- ph and Chloramines
- ph and Conductivity

```python
corrmatrix= df.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(corrmatrix,vmin=-0.2,vmax=1,annot=True,linewidth=.2,cmap='Blues')
```

```python
sns.pairplot(df)
plt.show()
```

```python
corrmatrix= df[['ph','Chloramines','Conductivity']].corr()
```

```python
corrmatrix
```

```python
plt.subplots(figsize=(20,8))
sns.heatmap(corrmatrix,vmin=-0.03,vmax=1,annot=True,linewidth=.2)
```

**Finding**
- **There is no varibles have either strong positive or strong negative correlation.**
- **Correlaion factor between ph and chloraamines is -0.025,this two variables have low negative correlation**
- **Correlaion factor between ph and conductivity is 0.014,this two variables have low positive correlation**
